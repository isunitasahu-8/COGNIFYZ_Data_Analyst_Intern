```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset.csv', encoding ='unicode_escape')
df.head()
```

```python
df.rename(columns ={'ï»¿Restaurant ID':'Restaurant ID'}, inplace = True)
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace =True)
df.isnull().sum()
```

```python
df.describe()
```

IDENTIFY THE CITY WITH HIGHEST NO. OF RESTAURANTS IN THE DATASET.

```python
df.columns
```

*Grouping the data by city and counting the number of restaurants in each city*

```python
city_restaurants_count = df.groupby('City')['Restaurant Name'].count()
city_restaurants_count
```

Identify the city with highest no of restaurant in it.

```python
city_with_most_restaurants = city_restaurants_count.idxmax()
city_with_most_restaurants
```

```python
number_Of_restaurants = city_restaurants_count.max()
number_Of_restaurants
```

Analyze the Average Rating for the Restaurant in each City.


```python
city_average_rating = df.groupby('City')['Aggregate rating'].mean()
city_average_rating
```

```python
print('The Average Rating for Restaurants in each city is as follows:')
print(city_average_rating)
```

```python
city_average_rating = df.groupby('City')['Aggregate rating'].mean().head(10)
```

```python
plt.figure(figsize=(14,6))
city_average_rating.plot(kind='bar', color = 'Pink', edgecolor ='Black')
plt.title('Avg Restro Rating in Each City')
plt.xlabel("Cities")
plt.ylabel("Ratings")
plt.xticks(rotation = 45)
plt.show()
```

CITY WITH THE HIGHEST AVERAGE RATING 

```python
city_with_highest_rating = city_average_rating.idxmax()
city_with_highest_rating
```

```python
Count_of_highest_rating = city_average_rating.max()
Count_of_highest_rating
```

```python
print(city_with_highest_rating,'has the highest rating of',Count_of_highest_rating )
```

SORTING THE CITIES BY THE AVERAGE RATING IN DESCENDING ORDER  
THEN SELECT TOP 3

```python
Top_three_avg_Rating_Cities = city_average_rating.sort_values(ascending=False).head(3)
Top_three_avg_Rating_Cities
```

```python
plt.figure(figsize= (14, 6))
Top_three_avg_Rating_Cities.plot(kind='bar', color = 'Pink', edgecolor ='Black')
plt.title("Top 3 Cities with the Most Ratings")
plt.xlabel("Cities")
plt.ylabel("Rating")
plt.xticks(rotation = 0)
plt.show()
```

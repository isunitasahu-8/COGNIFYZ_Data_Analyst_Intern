IDENTIFYING THE 3 TOP MOST COMMON CUISINES PRESENT IN THE DATASET:


```python
df['Cuisines'].value_counts()
```

```python
df['Cuisines'].value_counts().head(3)
```

```python
Top_three_Cuisines = df['Cuisines'].value_counts().head(3)
```

```python
plt.figure(figsize=(14,6))
Top_three_Cuisines.plot(kind ='bar', color ='skyblue', edgecolor ='black')
plt.title("Top Three Cuisines of the Restaurants")
plt.xlabel('Cuisines')
plt.ylabel("Count")
# For Better Readability
plt.xticks(rotation = 0)
plt.show()
```

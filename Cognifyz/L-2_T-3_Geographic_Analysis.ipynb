```python
map_restaurants = fl.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=5)
```

IDENTIFYING ANY PRESENT PATTERNS OR ANY PRESENT CLUSTERS OF RESTAURANTS IN SPECIFIC AREA.


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Now visualizing the data
plt.figure(figsize=(14,6))
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations')
plt.show()
```

```python
numeric_columns = df[['Latitude', 'Longitude']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_columns)
```

```python
wcss = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Using the elbow method [PLOT]
plt.figure(figsize=(14,6))
plt.plot(range(1, 7), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Using the elbow method [PlOT]
plt.figure(figsize=(14,6))
plt.plot(range(1, 7), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

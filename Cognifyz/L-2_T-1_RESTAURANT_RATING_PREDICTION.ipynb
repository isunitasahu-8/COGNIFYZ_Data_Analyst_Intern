ANALYZING THE DISTRIBUTION OF AGGREGATE RATINGS AND DETERMINING THE MOST COMMON RATING RANGE.


```python
df.columns
```

```python
# Ploting the Distribution of Aggregate Ratings
plt.figure(figsize=(14,6))
plt.hist(df['Aggregate rating'], bins = 5,color ='Pink', edgecolor='Black')
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Aggregate Ratings")
plt.xticks(rotation=0)
plt.show()
```

```python
# Determining the most common rating range
rating_counts = df['Aggregate rating'].value_counts()
most_common_rating = rating_counts.idxmax()
most_common_rating_count = rating_counts.max()
```

```python
# Calculating the average number of votes received by restaurants
average_votes = df['Votes'].mean()
print("Average number of votes received by restaurants:", average_votes)
```

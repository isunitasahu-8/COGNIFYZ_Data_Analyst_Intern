**AIM** : *The task is to determine the percentage of restaurants that offer online delivery and compare the average ratings of restaurants with and without online delivery.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset.csv", encoding ='unicode_escape')
df.head(10)
```

```python
df.rename(columns = {'ï»¿Restaurant ID':'Restaurant ID'}, inplace = True)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.shape
```

Identify the percentage of the restaurant that has online delivery service

```python
df.columns
```

```python
# Calculate the total number of restaurants
total_restaurants = len(df)
# Count the number of restaurants that offer online delivery
restaurants_with_delivery = df['Has Online delivery'].str.lower().eq('yes').sum()
# Calculate the percentage of restaurants that offer online delivery
percentage_with_delivery = (restaurants_with_delivery / total_restaurants) * 100
print(f"Percentage of restaurants that offer online delivery: {percentage_with_delivery:.2f}%")
```

```python
plt.figure(figsize=(14, 6))
plt.bar(['Restaurants with Online Delivery', 'Restaurants without Online Delivery'], 
        [percentage_with_delivery, 100 - percentage_with_delivery], 
        color=['#ffa07a', '#f08080'])  # Purple and light pink shades
plt.title('Percentage of Restaurants Offering Online Delivery')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.show()
```

NOW COMPARE THE AVERAGE RATINGS OF RESTAURANTS WITH AND WITHOUT ONLINE DELIVERY.


```python
restraurant_avg_rating = df.groupby('Has Online delivery')["Aggregate rating"].mean()
print("Avg Rating of Restaurants with and without Online Delivery:")
print(restraurant_avg_rating)
```

```python
# Set the figure size (width, height) in inches
plt.figure(figsize=(14,6))

# Plotting the bar chart with specified colors
plt.bar(['No', 'Yes'], restraurant_avg_rating, 
        color=['#FFDAB9', '#FFB6C1'])
plt.title("Average Ratings of Restaurants")
plt.xlabel("Restaurant With Online Delivery")
plt.ylabel("Average Rating")
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```

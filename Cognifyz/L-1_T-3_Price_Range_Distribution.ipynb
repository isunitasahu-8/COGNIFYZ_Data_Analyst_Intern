BUILD A HISTOGRAM OR BAR CHART TO VISUALIZE THE DITRIBUTION OF PRICE RANGES AMONG THE RESTAURANTS.

```python
plt.figure(figsize=(14,6))
df['Price range'].value_counts().sort_index().plot(kind='bar', color='Pink', edgecolor='Black')
plt.title("DITRIBUTION OF PRICE RANGES")
plt.xlabel("Price range")
plt.ylabel('Number of Restaurant')
plt.xticks(rotation=0)
plt.show()
```

CALCULATE THE PERCENTAGE OF RESTAURANTS IN EACH PRICE RANGE CATEGORY.


```python
price_range_counts = df['Price range'].value_counts()
price_range_counts
```

```python
plt.figure(figsize=(15,9))
explode = (0.01, 0.01, 0.01, 0.01)
colors = ['#ffb6c1', '#ffc0cb', '#ffa07a', '#f08080']
# Adjusted to shades of pink and subtle colors

price_range_counts.plot(kind='pie',labels=price_range_counts.index, autopct='%1.1f%%',explode=explode,colors=colors, startangle=140)
plt.title("PERCENTAGE OF RESTAURANTS IN EACH PRICE RANGE CATEGORY.")
plt.axis("equal")
plt.show()
```
